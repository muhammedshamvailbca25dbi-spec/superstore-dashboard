# Import Required Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

### Read the Superstore Dataset and Perform Initial Data Exploration

This step loads the dataset and checks:
- Dataset shape (rows and columns)
- Data types of each column
- Missing values
- Summary statistics for Sales, Quantity, Discount, and Profit


```python
df = pd.read_csv(r"C:\Users\msham\OneDrive\Documents\OJT\Project4.1\Data\samplesuperstore.csv")
print("Shape:", df.shape)
print(df.dtypes)
print(df.isnull().sum())
print(df[["Sales","Quantity","Discount","Profit"]].describe())
```

## Check for Duplicate Records

This step identifies:
- Total number of duplicate rows in the dataset
- Duplicate entries based on the Order ID column
- A preview of the duplicate records

```python
print("Duplicated rows:",df.duplicated().sum())
dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print(dupes.shape)
dupes.head()
```

## Calculate Shipping Time by Region

This step:
- Converts Order Date and Ship Date to datetime format
- Calculates the number of shipping days for each order
- Computes the average shipping time for each region

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
print(df.groupby("Region")["Shipping Days"].mean().round(1))
```

```python
df.dtypes
```

## Analyze Profit Margin by Sub-Category

This step:
- Calculates the profit margin percentage for each order
- Computes the average profit margin for each sub-category
- Sorts the results from lowest to highest profit margin


```python
df["Profit Margin %"] = (df["Profit"] / df["Sales"] * 100).round(2)

avg_margin = df.groupby("Sub-Category")["Profit Margin %"].mean().head(5).round(1)

avg_margin = avg_margin.sort_values()

print(avg_margin)
```

## Extract Order Year

This step:
- Extracts the year from the Order Date column
- Creates a new column named Order Year
- Displays all unique years present in the dataset

```python
df["Order Year"] = df["Order Date"].dt.year
print(df["Order Year"].unique())
```
